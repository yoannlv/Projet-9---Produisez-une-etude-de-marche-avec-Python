# Nettoyage des donnes

#### Manipulation dataframe Volaille_2021

```python
#Pivotage du dataframe
Volaille_2021 = pd.pivot_table(Volaille_2021, values = 'Valeur', index=['Zone'], columns = 'lment').reset_index()
```

```python
#Calcul de valeurs en milliers en units
Volaille_2021['Animaux Producteurs/Abattus'] = Volaille_2021['Animaux Producteurs/Abattus']*1000
Volaille_2021['Rserves'] = Volaille_2021['Rserves']*1000
```

```python
#Renommage et suppression colonnes
Volaille_2021 = Volaille_2021.rename(columns = {'Animaux Producteurs/Abattus' : 'Animaux Abattus', 'Rserves' : "Rserves (ttes)"})
Volaille_2021 = Volaille_2021.drop(['Rendement/Poids Carcasse'], axis=1)
Volaille_2021 = Volaille_2021.drop(['Production'], axis=1)
```

```python
Volaille_2021.head()
```

#### Manipulation dataframe Population

```python
#Changement chelle et type colonne Valeur
Population['Valeur'] = Population['Valeur']*1000
Population = Population.astype({'Valeur': int})
```

```python
#Renommage colonne
Population = Population.rename(columns = {'Valeur':'Population'})
```

```python
Population.head()
```

#### Manipulation dataframe StabPol

```python
#Renommage colonnes
StabPol = StabPol.rename(columns = {'Valeur' : 'Stabilit_Politique'})
```

#### Manipulation dataframe DispoAlim

#### Manipulation dataframe PIB

```python
#Renommage colonne
PIB = PIB.rename(columns = {'Valeur' : 'PIB/habt ($)'})
```

#### Manipulation dataframe Distance

```python
#Conservation distance par rapport  la France
Distance = Distance.loc[Distance['iso1']=='FRA']
#Renommage colonnnes
Distance = Distance.rename(columns={'iso2':'Code_pays', 'seadistance':'Distance'})
#Suppression colonne 'iso1'
Distance = Distance.drop(['iso1'], axis=1)
```

#### Manipulation dataframe Countries

```python
#Renommage colonne
Countries = Countries.rename(columns={'official_name_fr':'Zone','ISO3166-1-Alpha-3':'Code_pays'})
```

# Jointures

```python
#Jointure DispoAlim & Volaille_2021 + Prix_production
DispoAlim = DispoAlim.merge(Volaille_2021, how='left', on='Zone')
DispoAlim = DispoAlim.merge(Countries, how='left', on='Zone')
DispoAlim = DispoAlim.merge(Distance, how='left', on='Code_pays')
DispoAlim.head()
```

```python
#Jointure Population & StabPol
Pays = Population.merge(StabPol, how='left', on=('Zone', 'Anne'))
Pays = Pays.merge(PIB, how='left', on=('Zone', 'Anne'))
```

```python
Pays.head()
```

```python
#Cration dataframe moyenne
Pays_moy = Pays.groupby(by=['Zone']).mean()
```

```python
#Suppression colonne Anne
Pays_moy = Pays_moy.drop(['Anne'], axis=1)
```

```python
#Jointure DispoAlim & Pays_moy
Dataframe = DispoAlim.merge(Pays_moy, how='left', on='Zone')
```

```python
#Suppression colone Code_pays
Dataframe = Dataframe.drop(['Code_pays'], axis=1)
```

```python
Dataframe.head()
```

### Continent

```python
#Placement de la Chine en Asie
Dataframe['Continent'] = np.where(Dataframe['Zone']=='Chine - RAS de Hong-Kong', 'AS', Dataframe['Continent'])
Dataframe['Continent'] = np.where(Dataframe['Zone']=='Chine - RAS de Macao', 'AS', Dataframe['Continent'])
Dataframe['Continent'] = np.where(Dataframe['Zone']=='Chine, Taiwan Province de', 'AS', Dataframe['Continent'])
Dataframe['Continent'] = np.where(Dataframe['Zone']=='Chine, continentale', 'AS', Dataframe['Continent'])
#Affectation des valeurs nulles en Amrique du Nord
Dataframe['Continent'] = Dataframe['Continent'].replace(np.nan, 'NA')
```

```python
#Dtection des valeurs nulles
Dataframe.loc[Dataframe['Stabilit_Politique'].isnull()]
```

```python
#Imputation  la moyenne continentale de la stabilit politique
Dataframe.loc[(Dataframe['Stabilit_Politique'].isnull()) & (Dataframe['Continent']=='AS'),'Stabilit_Politique']=round(Dataframe['Stabilit_Politique'][Dataframe['Continent']=="AS"].mean(),2)
Dataframe.loc[(Dataframe['Stabilit_Politique'].isnull()) & (Dataframe['Continent']=='OC'),'Stabilit_Politique']=round(Dataframe['Stabilit_Politique'][Dataframe['Continent']=="OC"].mean(),2)
```

### Distance

```python
#Imputation  la moyenne continentale de la distance
Dataframe.loc[(Dataframe['Distance'].isnull()) & (Dataframe['Continent']=='AS'),'Distance']=round(Dataframe['Distance'][Dataframe['Continent']=="AS"].mean(),2)
Dataframe.loc[(Dataframe['Distance'].isnull()) & (Dataframe['Continent']=='AF'),'Distance']=round(Dataframe['Distance'][Dataframe['Continent']=="AF"].mean(),2)
Dataframe.loc[(Dataframe['Distance'].isnull()) & (Dataframe['Continent']=='EU'),'Distance']=round(Dataframe['Distance'][Dataframe['Continent']=="EU"].mean(),2)
Dataframe.loc[(Dataframe['Distance'].isnull()) & (Dataframe['Continent']=='NA'),'Distance']=round(Dataframe['Distance'][Dataframe['Continent']=="NA"].mean(),2)
Dataframe.loc[(Dataframe['Distance'].isnull()) & (Dataframe['Continent']=='SA'),'Distance']=round(Dataframe['Distance'][Dataframe['Continent']=="SA"].mean(),2)
Dataframe.loc[(Dataframe['Distance'].isnull()) & (Dataframe['Continent']=='OC'),'Distance']=round(Dataframe['Distance'][Dataframe['Continent']=="OC"].mean(),2)
```

### PIB

```python
#Dtection valeurs nulles
Dataframe.loc[Dataframe['PIB/habt ($)'].isnull()].head()
```

```python
#Imputation  la moyenne continentale du PIB/habt
Dataframe.loc[(Dataframe['PIB/habt ($)'].isnull()) & (Dataframe['Continent']=='AS'),'PIB/habt ($)']=round(Dataframe['PIB/habt ($)'][Dataframe['Continent']=="AS"].mean(),2)
Dataframe.loc[(Dataframe['PIB/habt ($)'].isnull()) & (Dataframe['Continent']=='AF'),'PIB/habt ($)']=round(Dataframe['PIB/habt ($)'][Dataframe['Continent']=="AF"].mean(),2)
Dataframe.loc[(Dataframe['PIB/habt ($)'].isnull()) & (Dataframe['Continent']=='EU'),'PIB/habt ($)']=round(Dataframe['PIB/habt ($)'][Dataframe['Continent']=="EU"].mean(),2)
Dataframe.loc[(Dataframe['PIB/habt ($)'].isnull()) & (Dataframe['Continent']=='NA'),'PIB/habt ($)']=round(Dataframe['PIB/habt ($)'][Dataframe['Continent']=="NA"].mean(),2)
Dataframe.loc[(Dataframe['PIB/habt ($)'].isnull()) & (Dataframe['Continent']=='SA'),'PIB/habt ($)']=round(Dataframe['PIB/habt ($)'][Dataframe['Continent']=="SA"].mean(),2)
Dataframe.loc[(Dataframe['PIB/habt ($)'].isnull()) & (Dataframe['Continent']=='OC'),'PIB/habt ($)']=round(Dataframe['PIB/habt ($)'][Dataframe['Continent']=="OC"].mean(),2)
```

```python
#Ajout colonne PIB total
Dataframe['PIB/habt ($)'] = Dataframe['PIB/habt ($)'].astype("float64")
Dataframe['PIB (M$)'] = Dataframe['Population']*Dataframe['PIB/habt ($)']
Dataframe['PIB (M$)'] = round(Dataframe['PIB (M$)']/1000000)
```

```python
#Arrondi colonne
Pays_moy['PIB/habt ($)'] = round(Pays_moy['PIB/habt ($)'],0)
Pays_moy['Population'] = round(Pays_moy['Population'],0)
Pays_moy['Stabilit_Politique'] = round(Pays_moy['Stabilit_Politique'],2)
```

### Animaux abattus et rserves

```python
#Imputation  la moyenne continentale du nombre d'animaux abattus
Dataframe.loc[(Dataframe['Animaux Abattus'].isnull()) & (Dataframe['Continent']=='AS'),'Animaux Abattus']=round(Dataframe['Animaux Abattus'][Dataframe['Continent']=="AS"].mean(),0)
Dataframe.loc[(Dataframe['Animaux Abattus'].isnull()) & (Dataframe['Continent']=='AF'),'Animaux Abattus']=round(Dataframe['Animaux Abattus'][Dataframe['Continent']=="AF"].mean(),0)
Dataframe.loc[(Dataframe['Animaux Abattus'].isnull()) & (Dataframe['Continent']=='EU'),'Animaux Abattus']=round(Dataframe['Animaux Abattus'][Dataframe['Continent']=="EU"].mean(),0)
Dataframe.loc[(Dataframe['Animaux Abattus'].isnull()) & (Dataframe['Continent']=='NA'),'Animaux Abattus']=round(Dataframe['Animaux Abattus'][Dataframe['Continent']=="NA"].mean(),0)
Dataframe.loc[(Dataframe['Animaux Abattus'].isnull()) & (Dataframe['Continent']=='SA'),'Animaux Abattus']=round(Dataframe['Animaux Abattus'][Dataframe['Continent']=="SA"].mean(),0)
Dataframe.loc[(Dataframe['Animaux Abattus'].isnull()) & (Dataframe['Continent']=='OC'),'Animaux Abattus']=round(Dataframe['Animaux Abattus'][Dataframe['Continent']=="OC"].mean(),0)
```

```python
#Imputation  la moyenne continentale de la rserve
Dataframe.loc[(Dataframe['Rserves (ttes)'].isnull()) & (Dataframe['Continent']=='AS'),'Rserves (ttes)']=round(Dataframe['Rserves (ttes)'][Dataframe['Continent']=="AS"].mean(),0)
Dataframe.loc[(Dataframe['Rserves (ttes)'].isnull()) & (Dataframe['Continent']=='AF'),'Rserves (ttes)']=round(Dataframe['Rserves (ttes)'][Dataframe['Continent']=="AF"].mean(),0)
Dataframe.loc[(Dataframe['Rserves (ttes)'].isnull()) & (Dataframe['Continent']=='EU'),'Rserves (ttes)']=round(Dataframe['Rserves (ttes)'][Dataframe['Continent']=="EU"].mean(),0)
Dataframe.loc[(Dataframe['Rserves (ttes)'].isnull()) & (Dataframe['Continent']=='NA'),'Rserves (ttes)']=round(Dataframe['Rserves (ttes)'][Dataframe['Continent']=="NA"].mean(),0)
Dataframe.loc[(Dataframe['Rserves (ttes)'].isnull()) & (Dataframe['Continent']=='SA'),'Rserves (ttes)']=round(Dataframe['Rserves (ttes)'][Dataframe['Continent']=="SA"].mean(),0)
Dataframe.loc[(Dataframe['Rserves (ttes)'].isnull()) & (Dataframe['Continent']=='OC'),'Rserves (ttes)']=round(Dataframe['Rserves (ttes)'][Dataframe['Continent']=="OC"].mean(),0)
```

```python
#Conversion type de valeurs
Dataframe['Animaux Abattus'] = Dataframe['Animaux Abattus'].astype("Int64")
Dataframe['Rserves (ttes)'] = Dataframe['Rserves (ttes)'].astype("Int64")
```
